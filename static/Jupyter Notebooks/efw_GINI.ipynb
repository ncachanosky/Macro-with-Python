```python
"|IMPORT PACKAGES|"
import pandas             as pd
import datetime
from   bokeh.plotting     import show, figure, output_file, save
from   bokeh.io           import show, output_notebook, curdoc, export_png
from   bokeh.models       import ColumnDataSource,LinearAxis, Range1d, NumeralTickFormatter, LabelSet, Label, BoxAnnotation, DatetimeTickFormatter
from   bokeh.models.tools import HoverTool
from   bokeh.transform    import dodge
```

```python
"|IMPORT DATA|"
path = r'https://github.com/ncachanosky/research/blob/master/Economic%20Series/'
file = r'Resumen%20Estadistico%20-%20Internacional.xlsx?raw=true'

IO   = path + file

sheet = 'DATA'

data = pd.read_excel(IO, sheet_name = sheet, usecols="A:F,L,Q", skiprows=2, engine='openpyxl') # Be patient...

data = data.rename(columns={"GDP (PPP)":"GDP"})
data = data.rename(columns={"GINI (WB)":"GINI"})
data = data[data.YEAR == 2018].reset_index()
data = data.drop(['index'], axis = 1)
```

```python
"|CHECK DATA|"
data
```

```python
"|BUILD PLOT|"

X = ['Menor libertad económica',
     '3er cuartil'             ,
     '2do cuartil'             ,
     'Mayor libertad económica']

Q1 = data[data.QUART == 1]
Q2 = data[data.QUART == 2]
Q3 = data[data.QUART == 3]
Q4 = data[data.QUART == 4]

BAR1 = Q1['GINI'].mean()
BAR2 = Q2['GINI'].mean()
BAR3 = Q3['GINI'].mean()
BAR4 = Q4['GINI'].mean()

BAR = [BAR1, BAR2, BAR3, BAR4]

colors = ['red', 'gold', 'green', 'blue']

#BUILD FIGURE
p = figure(title        = "EL HUB ECONÓMICO | LIBERTAD ECONÓMICA E ÍNDICE DE GINI",
           x_axis_label = "",
           y_axis_label = "% del ingreso total",
           y_range      = (0, 60),
           x_range      = X,
           plot_height  = 400,
           plot_width   = 700)

p.toolbar_location = "right"
p.toolbar.autohide = True

#AXIS 1 (LEFT)
p.vbar(x=X            ,
       top=BAR        ,
       color=colors   ,
       width=0.5      ,
       fill_alpha=0.5)

#LEGEND
show(p)
```

```python
"|EXPORT .PNG FILE|"

export_png(p, filename="efw_GINI.png")
```

```python
"|# CREATE HTML FILE|"

output_file(filename="efw_GINI.html", title="EFW y GINI")
save(p)
```

```python
"|CREATE JSON FILE|"
import json
import bokeh.embed

from bokeh.embed import json_item

j = json.dumps(json_item(p, "efw_GINI"))
with open("efw_GINI.json", "w") as fp:
    json.dump(j, fp)
```
