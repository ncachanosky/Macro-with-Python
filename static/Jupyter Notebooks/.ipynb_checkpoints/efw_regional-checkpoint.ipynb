```python
"|IMPORT PACKAGES|"
import numpy              as np
import pandas             as pd
import datetime
from   math               import pi
from   bokeh.plotting     import show, figure, output_file, save
from   bokeh.io           import show, output_notebook, curdoc, export_png
from   bokeh.models       import ColumnDataSource,LinearAxis, Range1d, NumeralTickFormatter, LabelSet, Label, BoxAnnotation, DatetimeTickFormatter, Text, Span
from   bokeh.models.tools import HoverTool
from   bokeh.models       import Arrow, NormalHead, OpenHead, VeeHead
from   bokeh.transform    import dodge
from   datetime           import datetime as dt
```

```python
"|IMPORT DATA|"
path = r'https://github.com/ncachanosky/research/blob/master/Economic%20Series/'
file = r'Resumen%20Estadistico%20-%20Internacional.xlsx?raw=true'
#file = r'Resumen%20Estadistico%20-%20Argentina.xlsx'
IO   = path + file

sheet = 'DATA'

data = pd.read_excel(IO, sheet_name = sheet, usecols="A:E,K", skiprows=2, engine='openpyxl') # Be patient...
```

```python
data
```

```python
"\BUILD REGIONAL SERIES|"
year_start = 1970
year_end   = 2018

YEAR = np.zeros(year_end+1 - year_start)

EFW_WORLD = np.zeros(year_end+1 - year_start)
EFW_AF    = np.zeros(year_end+1 - year_start)
EFW_LA    = np.zeros(year_end+1 - year_start)
EFW_NA    = np.zeros(year_end+1 - year_start)
EFW_AS    = np.zeros(year_end+1 - year_start)
EFW_EU    = np.zeros(year_end+1 - year_start)
EFW_OC    = np.zeros(year_end+1 - year_start)
EFW_ARG   = np.zeros(year_end+1 - year_start)

i = 0
while year_start < year_end+1:
    YEAR[i] = year_start
        
    S = data[data.YEAR==year_start]
    S['NUM'] = S['EFW']*S['POPULATION']
        
    
    S_AF = S[S.REGION=="Africa"]
    S_AF['NUM'] = S_AF['EFW']*S_AF['POPULATION']
    
    S_LA = S[S.REGION=="Latin America"]
    S_LA['NUM'] = S_LA['EFW']*S_LA['POPULATION']
    
    S_NA = S[S.REGION=="North America"]
    S_NA['NUM'] = S_NA['EFW']*S_NA['POPULATION']
    
    S_AS = S[S.REGION=="Asia"]
    S_AS['NUM'] = S_AS['EFW']*S_AS['POPULATION']
    
    S_EU = S[S.REGION=="Europe"]
    S_EU['NUM'] = S_EU['EFW']*S_EU['POPULATION']
    
    S_OC = S[S.REGION=="Oceania"]
    S_OC['NUM'] = S_OC['EFW']*S_EU['POPULATION']
    
    
    EFW_WORLD[i] = S['NUM'].sum() / S['POPULATION'].sum()
    EFW_AF[i]    = S_AF['NUM'].sum() / S_AF['POPULATION'].sum()
    EFW_LA[i]    = S_LA['NUM'].sum() / S_LA['POPULATION'].sum()
    EFW_NA[i]    = S_NA['NUM'].sum() / S_NA['POPULATION'].sum()
    EFW_AS[i]    = S_AS['NUM'].sum() / S_AS['POPULATION'].sum()
    EFW_EU[i]    = S_EU['NUM'].sum() / S_EU['POPULATION'].sum()
    EFW_OC[i]    = S_OC['NUM'].sum() / S_OC['POPULATION'].sum()
    
    S_ARG = S[S.COUNTRY=="Argentina"]
    EFW_ARG[i]   = S_ARG['EFW']
        
    i = i+1
    year_start = year_start + 1

    
dataset = np.vstack([YEAR, EFW_WORLD, EFW_AF, EFW_LA, EFW_NA, EFW_AS, EFW_EU, EFW_OC, EFW_ARG])
dataset = np.transpose(dataset)

data = pd.DataFrame(dataset)
data.replace(0, np.nan, inplace=True)
data.columns = ['YEAR'    ,
                'WORLD'   ,
                'AFRICA'  ,
                'LATIN_AM',
                'NORTH_AM',
                'ASIA'    ,
                'EUROPE'  ,
                'OCEANIA' ,
                'ARGENTINA']

data['WORLD2']     = data['WORLD'].interpolate()
data['AFRICA2']    = data['AFRICA'].interpolate()
data['LATIN_AM2']  = data['LATIN_AM'].interpolate()
data['NORTH_AM2']  = data['NORTH_AM'].interpolate()
data['ASIA2']      = data['ASIA'].interpolate()
data['EUROPE2']    = data['EUROPE'].interpolate()
data['OCEANIA2']   = data['OCEANIA'].interpolate()
data['ARGENTINA2'] = data['ARGENTINA'].interpolate()
```
