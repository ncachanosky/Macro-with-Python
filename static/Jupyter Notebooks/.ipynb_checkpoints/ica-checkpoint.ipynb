```python
"|IMPORT PACKAGES|"
import pandas             as pd
import datetime
from   bokeh.plotting     import show, figure, output_file, save
from   bokeh.io           import show, output_notebook, curdoc, export_png
from   bokeh.models       import ColumnDataSource,LinearAxis, Range1d, NumeralTickFormatter, LabelSet, Label, Legend
from   bokeh.models       import BoxAnnotation, DatetimeTickFormatter, Span
from   bokeh.models.tools import HoverTool
```

```python
"|IMPORT DATA|"
path = r'https://github.com/ncachanosky/research/blob/master/Economic%20Series/'
file = r'Resumen%20Estadistico%20-%20Argentina.xlsx?raw=true'
IO   = path + file

sheet = 'ICA'

data = pd.read_excel(IO, sheet_name = sheet, usecols="A:G,O:R", skiprows=6, engine='openpyxl') # Be patient...

data = data.rename(columns={"Fecha"                        :"YEAR"})
data = data.rename(columns={"Expo Totales"                 :"EXPO"})
data = data.rename(columns={"Combustibles y Energía"       :"COMB"})
data = data.rename(columns={"Impo Totales"                 :"IMPO"})
data = data.rename(columns={"G1 B. capital + Partes y pzas":"G1"})
data = data.rename(columns={"G2 B.Intermedios +Combust."   :"G2"})
data = data.rename(columns={"G3 B.Consumo+Vehículos+Resto" :"G3"})
data = data.rename(columns={"Saldo Comercial"              :"SALDO"})

data['G1'] = -data['G1']
data['G2'] = -data['G2']
data['G3'] = -data['G3']
```

```python
"|CHECK DATA|"
data.tail(25)
```

```python
"|BUILD PLOT|"

cds = ColumnDataSource(data)

colors_X  = ['midnightblue','blue','steelblue','lightblue']
legends_X = ['PP','MOA','MOI','Comb. y energía']

colors_Z  = ['darkred','firebrick','indianred']
legends_Z = ['Bs. de capital + partes y piezas','Bs. intermedios y comb.','Consumo, vehiculos, y resto']

#BUILD FIGURE
p = figure(title        = "INTERCAMBIO COMERCIAL ARGENTINO (ICA)",
           x_range      = (1992, 2021),
           y_range      = (-100000, 100000),
           y_axis_label = "Millones de dólares", 
           plot_height  = 400,
           plot_width   = 700)

p.toolbar_location = "above"
p.toolbar.autohide = True

#AXIS 1 (LEFT)
p0 = p.line('YEAR', 'SALDO', line_color='black', width=2, alpha=0.75, muted_alpha=0.75, source=cds)

p1 = p.varea_stack(['PP','MOA','MOI','COMB'], x='YEAR', color=colors_X, fill_alpha=0.50, muted_alpha=0.2, source=cds)
p2 = p.vline_stack(['PP','MOA','MOI','COMB'], x='YEAR', color=colors_X, line_alpha=0.75, width=1, source=cds)

p3 = p.varea_stack(['G1','G2','G3'], x='YEAR', color=colors_Z, fill_alpha=0.50, muted_alpha=0.2, source=cds)
p4 = p.vline_stack(['G1','G2','G3'], x='YEAR', color=colors_Z, line_alpha=0.75, width=1, source=cds)

#p.add_tools(HoverTool(renderers=[p1], tooltips = [("PP"             , "@PP")]  , mode="vline"))
#p.add_tools(HoverTool(renderers=[p2], tooltips = [("MOA"            , "@MOI")] , mode="vline"))
#p.add_tools(HoverTool(renderers=[p3], tooltips = [("MOI"            , "@MOA")] , mode="vline"))
#p.add_tools(HoverTool(renderers=[p4], tooltips = [("Comb. y energía", "@COMB")], mode="vline"))

#date1 = Span(location=1984, dimension="height", line_color="black", line_dash="dashed", line_alpha=0.5)
#date2 = Span(location=1989, dimension="height", line_color="black", line_dash="dashed", line_alpha=0.5)
#date3 = Span(location=1999, dimension="height", line_color="black", line_dash="dashed", line_alpha=0.5)
#date4 = Span(location=2001, dimension="height", line_color="black", line_dash="dashed", line_alpha=0.5)
#date5 = Span(location=2007, dimension="height", line_color="black", line_dash="dashed", line_alpha=0.5)
#date6 = Span(location=2015, dimension="height", line_color="black", line_dash="dashed", line_alpha=0.5)

#p.add_layout(date1)
#p.add_layout(date2)
#p.add_layout(date3)
#p.add_layout(date4)
#p.add_layout(date5)
#p.add_layout(date6)

p.yaxis[0].formatter = NumeralTickFormatter(format="$0,0")

h_axis = Span(location=0, dimension="width", line_color="black")
p.add_layout(h_axis)

legend = Legend(items=[
    ("PP"                        , [p1[0]]),
    ("MOA"                       , [p1[1]]),
    ("MOI"                       , [p1[2]]),
    ("Comb. y energía"           , [p1[3]]),
    ("Capital, partes y piezas" , [p3[0]]),
    ("Interm. y comb."         , [p3[1]]),
    ("Cons., veh. y resto", [p3[2]]) ], location="center", click_policy="mute", orientation="vertical")

p.add_layout(legend, 'right')


#p.legend.location     = "below"
#p.legend.orientation  = "horizontal"
#p.legend.click_policy = "mute"
show(p)
```

```python
"|EXPORT .PNG FILE|"

export_png(p, filename="ica.png")
```

```python
"|# CREATE HTML FILE|"

output_file(filename="ica.html", title="Intercambio Comercial Argentino")
save(p)
```
