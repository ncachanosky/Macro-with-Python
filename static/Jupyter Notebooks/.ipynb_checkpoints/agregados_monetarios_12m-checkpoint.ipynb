```python
"|IMPORT PACKAGES|"
import numpy              as np
import pandas             as pd
import datetime
from   bokeh.plotting     import show, figure, output_file, save
from   bokeh.io           import show, output_notebook, curdoc, export_png
from   bokeh.models       import ColumnDataSource,LinearAxis, Range1d, NumeralTickFormatter, LabelSet, Label, BoxAnnotation, DatetimeTickFormatter, Text, Span
from   bokeh.models.tools import HoverTool
from   bokeh.models       import Arrow, NormalHead, OpenHead, VeeHead
from   bokeh.transform    import dodge
from   datetime           import datetime as dt
```

```python
"|IMPORT DATA|"
path = r'https://github.com/ncachanosky/research/blob/master/Economic%20Series/'
file = r'Resumen%20Estadistico%20-%20Argentina.xlsx?raw=true'
IO   = path + file

sheet = 'MONTHLY'

data = pd.read_excel(IO, sheet_name = sheet, usecols="B,AZ,BU", skiprows=2, nrows=399, engine='openpyxl') # Be patient...

data = data.rename(columns={"Monetary Base (%) [12 meses]":"BM"})
data = data.rename(columns={"M2 (%) [12M]"                :"M2"})

data['BM'] = data['BM']*100
data['M2'] = data['M2']*100
```

```python
"|CHECK DATA|"
data
```

```python
"|BUILD PLOT|"

cds = ColumnDataSource(data)

#BCRA Presidents shades and names
PratGay     = BoxAnnotation(left=dt(2002,12,31).timestamp()*1000, right=dt(2004, 9,23).timestamp()*1000, line_color="black", fill_color="skyblue", fill_alpha=0.1)
Redrado     = BoxAnnotation(left=dt(2004, 9,24).timestamp()*1000, right=dt(2010, 1,10).timestamp()*1000, line_color="black", fill_color="purple" , fill_alpha=0.1)
delPont     = BoxAnnotation(left=dt(2010, 1,23).timestamp()*1000, right=dt(2013,11,18).timestamp()*1000, line_color="black", fill_color="green"  , fill_alpha=0.1)
Fabrega     = BoxAnnotation(left=dt(2013,11,19).timestamp()*1000, right=dt(2014,10, 1).timestamp()*1000, line_color="black", fill_color="orange" , fill_alpha=0.1)
Vanoli      = BoxAnnotation(left=dt(2014,10, 2).timestamp()*1000, right=dt(2015,12,10).timestamp()*1000, line_color="black", fill_color="black"  , fill_alpha=0.1)
Sturz       = BoxAnnotation(left=dt(2015,12,11).timestamp()*1000, right=dt(2018, 6,14).timestamp()*1000, line_color="black", fill_color="gold"   , fill_alpha=0.1)
Caputo      = BoxAnnotation(left=dt(2018, 6,15).timestamp()*1000, right=dt(2018, 9,25).timestamp()*1000, line_color="black", fill_color="gray"   , fill_alpha=0.1)
Sandleris   = BoxAnnotation(left=dt(2018, 9,26).timestamp()*1000, right=dt(2019,12,10).timestamp()*1000, line_color="black", fill_color="maroon" , fill_alpha=0.1)

PratGay_N   = Label(x=dt(2003,11, 2).timestamp()*1000, y=140, text="Prat-Gay"      , text_font_size="8pt", text_align="center", angle_units="deg", angle= 0)
Redrado_N   = Label(x=dt(2007, 5,25).timestamp()*1000, y=140, text="Redrado"       , text_font_size="8pt", text_align="center", angle_units="deg", angle= 0)
delPont_N   = Label(x=dt(2011,12,11).timestamp()*1000, y=140, text="Marcó del Pont", text_font_size="8pt", text_align="center", angle_units="deg", angle= 0)
Fabrega_N   = Label(x=dt(2014, 6,26).timestamp()*1000, y=145, text="Fábrega"       , text_font_size="8pt", text_align="right" , angle_units="deg", angle=90)
Vanoli_N    = Label(x=dt(2015, 7, 7).timestamp()*1000, y=145, text="Vanoli"        , text_font_size="8pt", text_align="right" , angle_units="deg", angle=90)
Sturz_N     = Label(x=dt(2017, 5,12).timestamp()*1000, y=145, text="Sturzenneger"  , text_font_size="8pt", text_align="right" , angle_units="deg", angle=90)
Caputo_N    = Label(x=dt(2018,10, 4).timestamp()*1000, y=145, text="Caputo"        , text_font_size="8pt", text_align="right" , angle_units="deg", angle=90)
Sandleris_N = Label(x=dt(2019, 7, 3).timestamp()*1000, y=145, text="Sandleris"     , text_font_size="8pt", text_align="right" , angle_units="deg", angle=90)

#BUILD FIGURE
p = figure(title = "EL HUB ECONÓMICO | VARIACIÓN INTERANUAL: BM Y M2",
           x_range      = (dt(1999, 1,31).timestamp()*1000, dt(2021, 3,31).timestamp()*1000),
           x_axis_label = "",
           y_axis_label = "",
           x_axis_type  = "datetime",
           plot_height  = 400,
           plot_width   = 700)

p.toolbar_location = "right"
p.toolbar.autohide = True

#AXIS 1 (LEFT)
p1 = p.line('MONTH', 'BM', color='black', line_alpha=0.75, width=2, legend_label='BM', muted_alpha=0.2, source=cds)
p2 = p.line('MONTH', 'M2', color='red'  , line_alpha=0.75, width=2, legend_label='M2', muted_alpha=0.2, source=cds)

p.add_tools(HoverTool(renderers=[p1], tooltips = [("BM", "@BM")], mode="vline"))
p.add_tools(HoverTool(renderers=[p2], tooltips = [("M2", "@M2")], mode="vline"))

#BCRA PRESIDENTS
p.add_layout(PratGay)
p.add_layout(Redrado)
p.add_layout(delPont)
p.add_layout(Fabrega)
p.add_layout(Vanoli)
p.add_layout(Sturz)
p.add_layout(Caputo)
p.add_layout(Sandleris)

p.add_layout(PratGay_N)
p.add_layout(Redrado_N)
p.add_layout(delPont_N)
p.add_layout(Fabrega_N)
p.add_layout(Vanoli_N)
p.add_layout(Sturz_N)
p.add_layout(Caputo_N)
p.add_layout(Sandleris_N)

#LEGEND
h_axis = Span(location=0, dimension="width", line_color="black")
p.add_layout(h_axis)
p.xaxis[0].formatter = DatetimeTickFormatter(months='%b %Y')
p.legend.location     = "top_left"
p.legend.orientation  = "vertical"
p.legend.click_policy = "mute"
show(p)
```

```python
"|EXPORT .PNG FILE|"

export_png(p, filename="agregados_monetarios_12m.png")
```

```python
"|# CREATE HTML FILE|"

output_file(filename="agregados_monetarios_12m.html", title="Agregados Monetarios: 12M")
save(p)
```

```python
"|CREATE JSON FILE|"
import json
import bokeh.embed

from bokeh.embed import json_item

j = json.dumps(json_item(p, "agregados_monetarios_12m"))
with open("agregados_monetarios_12m.json", "w") as fp:
    json.dump(j, fp)
```
