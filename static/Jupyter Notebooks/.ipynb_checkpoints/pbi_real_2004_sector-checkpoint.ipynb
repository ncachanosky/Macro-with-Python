```python
"|IMPORT PACKAGES|"
import pandas             as pd
import datetime
from   bokeh.plotting     import show, figure, output_file, save
from   bokeh.io           import show, output_notebook, curdoc, export_png
from   bokeh.models       import ColumnDataSource,LinearAxis, Range1d, NumeralTickFormatter, LabelSet, Label, BoxAnnotation, DatetimeTickFormatter, Legend, LegendItem
from   bokeh.models.tools import HoverTool
```

```python
"|IMPORT DATA|"
path = r'https://github.com/ncachanosky/research/blob/master/Economic%20Series/'
file = r'Resumen%20Estadistico%20-%20Argentina.xlsx?raw=true'
IO   = path + file

sheet = 'PBI'

data = pd.read_excel(IO, sheet_name = sheet, usecols="HR:IO", skiprows=6, nrows=17, engine='openpyxl') # Be patient...

data.columns = ['YEAR',
                'A','B','C','D','E','F',
                'BIENES',
                'G','H','I','J','K','L','M','N','O','P',
                'SERVICIOS','VA BRUTO','IVA','IMP IMPORTACIONES','IMP NETOS','PBI']
```

```python
"|CHECK DATA|"         
data
```

```python
"|BUILD PLOT|"

cds = ColumnDataSource(data)

colors=['black',     # 1
        'blue' ,     # 2
        'red'  ,     # 3
        'green',     # 4
        'slateblue', # 5
        'teal',      # 6
        'tan',       # 7
        'plum',      # 8
        'salmon',    # 9
        'orange',    #10
        'olive',     #11
        'maroon',    #12
        'lime',      #13
        'gold',      #14
        'peru',      #15
        'firebrick'] #16

legends=['Agricultura',                        # 1
         'Pesca',                              # 2
         'Minería',                            # 3
         'Manufacturas',                       # 4
         'Electricidad, gas, y agua',          # 5
         'Construcción',                       # 6
         'Comercio',                           # 7
         'Hoteles y restaurantes',             # 8
         'Transp., almacenamiento y comunic.', # 9
         'Intermediación financiera',          #10
         'Sector inmobiliario',                #11
         'Adm. pública y defensa',             #12
         'Educación',                          #13
         'Serv. sociales y salud',             #14
         'Otros',                              #15
         'Servicios domésticos']               #16
        
#BUILD FIGURE
p = figure(title = "EL HUB ECONÓMICO | PBI REAL (2004) POR SECTOR: VALOR AGREGADO BRUTO",
           y_axis_label = "Millones de pesos", 
           plot_height  = 600,
           plot_width   = 700)

p.toolbar_location = "right"
p.toolbar.autohide = True
'G','H','I','J','K','L','M','N','O','P'
#AXIS 1 (LEFT)
p1 = p.varea_stack(['A','B','C','D','E','F','G','H','I','J','K','L','M','N','O','P'], x='YEAR', color=colors, fill_alpha=0.5, muted_alpha=0.20, source=cds)
p2 = p.vline_stack(['A','B','C','D','E','F','G','H','I','J','K','L','M','N','O','P'], x='YEAR', color=colors, line_alpha=1.0, muted_alpha=0.20, source=cds)       

legend1 = Legend(items=[(legends[ 0], [p1[ 0]]),
                        (legends[ 1], [p1[ 1]]),
                        (legends[ 2], [p1[ 2]]),
                        (legends[ 3], [p1[ 3]]),
                        (legends[ 4], [p1[ 4]]),
                        (legends[ 5], [p1[ 5]])], orientation="horizontal", click_policy='mute')
                       
legend2 = Legend(items=[(legends[ 6], [p1[ 6]]),
                        (legends[ 7], [p1[ 7]]),
                        (legends[ 8], [p1[ 8]]),
                        (legends[ 9], [p1[ 9]]) ], orientation="horizontal", click_policy='mute')
                
legend3 = Legend(items=[(legends[10], [p1[10]]),
                        (legends[11], [p1[11]]),
                        (legends[12], [p1[12]]),
                        (legends[13], [p1[13]]) ], orientation="horizontal", click_policy='mute')
        
legend4 = Legend(items=[(legends[14], [p1[14]]),
                        (legends[15], [p1[15]]) ], orientation="horizontal", click_policy='mute')


p.add_layout(legend1, "below")
p.add_layout(legend2, "below")
p.add_layout(legend3, "below")
p.add_layout(legend4, "below")

p.yaxis[0].formatter = NumeralTickFormatter(format="$0,0")

show(p)
```

```python
"|EXPORT .PNG FILE|"

export_png(p, filename="pbi_real_2004_sector.png")
```

```python
"|# CREATE HTML FILE|"

output_file(filename="pbi_real_2004_sector.html", title="PBI real 2004 (sector)")
save(p)
```

```python
"|CREATE JSON FILE|"
import json
import bokeh.embed

from bokeh.embed import json_item

j = json.dumps(json_item(p, "pbi_real_2004_sector"))
with open("pbi_real_2004_sector.json", "w") as fp:
    json.dump(j, fp)
```

```python
"|CREATE SECTOR SHARES|"

data['TOTAL'] = data['A'] + data['B'] + data['C'] + data['D'] + data['E'] + data['F'] + data['G'] + data['H'] + \
                data['I'] + data['J'] + data['K'] + data['L'] + data['M'] + data['N'] + data['O'] + data['P']

data['A'] = data['A']/data['TOTAL'] * 100
data['B'] = data['B']/data['TOTAL'] * 100
data['C'] = data['C']/data['TOTAL'] * 100
data['D'] = data['D']/data['TOTAL'] * 100
data['E'] = data['E']/data['TOTAL'] * 100
data['F'] = data['F']/data['TOTAL'] * 100
data['G'] = data['G']/data['TOTAL'] * 100
data['H'] = data['H']/data['TOTAL'] * 100
data['I'] = data['I']/data['TOTAL'] * 100
data['J'] = data['J']/data['TOTAL'] * 100
data['K'] = data['K']/data['TOTAL'] * 100
data['L'] = data['L']/data['TOTAL'] * 100
data['M'] = data['M']/data['TOTAL'] * 100
data['N'] = data['N']/data['TOTAL'] * 100
data['O'] = data['O']/data['TOTAL'] * 100
data['P'] = data['P']/data['TOTAL'] * 100

data
```
