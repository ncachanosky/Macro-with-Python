```python
"|IMPORT PACKAGES|"
import numpy              as np
import pandas             as pd
import datetime
from   bokeh.plotting     import show, figure, output_file, save
from   bokeh.io           import show, output_notebook, curdoc, export_png
from   bokeh.models       import ColumnDataSource,LinearAxis, Range1d, NumeralTickFormatter, LabelSet, Label, BoxAnnotation, DatetimeTickFormatter, Text, Span
from   bokeh.models.tools import HoverTool
from   bokeh.models       import Arrow, NormalHead, OpenHead, VeeHead
from   bokeh.transform    import dodge
from   datetime           import datetime as dt
```

```python
"|IMPORT DATA|"
path = r'https://github.com/ncachanosky/research/blob/master/Economic%20Series/'
file = r'Resumen%20Estadistico%20-%20Argentina.xlsx?raw=true'
IO   = path + file

sheet = 'DEFICIT FINANCIERO'

data = pd.read_excel(IO, sheet_name = sheet, usecols="AD,AG:AH", nrows=65, engine='openpyxl') # Be patient...

data = data[39:]
```

```python
"|CHECK DATA|"
data
```

```python
"|BUILD PLOT|"

cds = ColumnDataSource(data)

#BUILD FIGURE
p = figure(title        = 'EL HUB ECONÓMICO | MINISTROS DE ECONOMÍA (DÍAS EN EJERCICIO)',
           x_axis_label = '',
           y_axis_label = 'Dias',
           x_range      = data['MINISTRO'],
           y_range      = (0, 2500),
           plot_height  = 500,
           plot_width   = 700)

p.toolbar_location = "right"
p.toolbar.autohide = True

#AXIS 1 (LEFT)
p1 = p.vbar(x='MINISTRO', bottom=0, top='DAYS', color='blue', width=0.5, fill_alpha=0.50, legend_label='Días por ministro', muted_alpha=0.2, source=cds)
p2 = p.line('MINISTRO', 'AVG', color='red', line_alpha=0.5, muted_alpha=0.2, legend_label='Dias promedio', line_dash='dashed', source=cds)

p.add_tools(HoverTool(renderers=[p1], tooltips = [('Días'         , '@DAYS{0}')], mode='vline'))
p.add_tools(HoverTool(renderers=[p2], tooltips = [('Días promedio', '@AVG{0}' )], mode='vline'))

#LEGEND
p.xaxis.major_label_orientation = 1.2
p.legend.location     = "top_left"
p.legend.orientation  = "vertical"
p.legend.click_policy = "mute"
show(p)
```

```python
"|EXPORT .PNG FILE|"

export_png(p, filename="ministros_dias.png")
```

```python
"|# CREATE HTML FILE|"

output_file(filename="ministros_dias.html", title="Ministros de Economia")
save(p)
```

```python
"|CREATE JSON FILE|"
import json
import bokeh.embed

from bokeh.embed import json_item

j = json.dumps(json_item(p, "ministros_dias"))
with open("ministros_dias.json", "w") as fp:
    json.dump(j, fp)
```
