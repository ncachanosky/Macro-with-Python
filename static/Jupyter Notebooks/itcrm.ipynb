```python
"|IMPORT PACKAGES|"
import pandas             as pd
import datetime
from   bokeh.plotting     import show, figure, output_file, save
from   bokeh.io           import show, output_notebook, curdoc, export_png
from   bokeh.models       import ColumnDataSource,LinearAxis, Range1d, NumeralTickFormatter, LabelSet, Label, BoxAnnotation, DatetimeTickFormatter, Legend
from   bokeh.models.tools import HoverTool
from   bokeh.palettes     import Category20
from   bokeh.models       import Arrow, NormalHead, OpenHead, VeeHead
from   datetime           import datetime as dt
```

```python
"|IMPORT DATA|"
path = r'https://github.com/ncachanosky/research/blob/master/Economic%20Series/'
file = r'Resumen%20Estadistico%20-%20Argentina.xlsx?raw=true'
IO   = path + file

sheet = 'ITCRM'

data = pd.read_excel(IO, sheet_name = sheet, usecols="A:P", skiprows=1, engine='openpyxl') # Be patient...

data = data.rename(columns={"Período"             :"MONTH"})
data = data.rename(columns={"ITCRM "              :"ITCRM"})
data = data.rename(columns={"ITCRB Brasil"        :"BRASIL"})
data = data.rename(columns={"ITCRB Canadá"        :"CANADA"})
data = data.rename(columns={"ITCRB Chile"         :"CHILE"})
data = data.rename(columns={"ITCRB Estados Unidos":"EEUU"})
data = data.rename(columns={"ITCRB México"        :"MEXICO"})
data = data.rename(columns={"ITCRB Uruguay"       :"URUGUAY"})
data = data.rename(columns={"ITCRB China"         :"CHINA"})
data = data.rename(columns={"ITCRB India"         :"INDIA"})
data = data.rename(columns={"ITCRB Japón"         :"JAPON"})
data = data.rename(columns={"ITCRB Reino Unido"   :"UK"})
data = data.rename(columns={"ITCRB Suiza"         :"SUIZA"})
data = data.rename(columns={"ITCRB Zona Euro"     :"EURO"})
data = data.rename(columns={"ITCRB Vietnam"       :"VIETNAM"})
data = data.rename(columns={"ITCRB Sudamérica*"   :"LATAM"})
```

```python
"|CHECK DATA|"
data
```

```python
"|BUILD PLOT|"

cds    = ColumnDataSource(data)

#BUILD FIGURE
p = figure(title        = 'EL HUB ECONÓMICO | INDICE DE TIPO DE CAMBIO REAL MULTILATERAL',
           x_axis_label = '',
           y_axis_label = '17-Ene-15 = 100',
           y_range      = (0, 400),
           plot_height  = 400,
           plot_width   = 700)

p.xaxis[0].formatter = DatetimeTickFormatter(months="%b %Y")
p.toolbar_location = "right"
p.toolbar.autohide = True

#AXIS 1 (LEFT)
p1  = p.line("MONTH", "ITCRM"  , line_alpha=0.75, width=4, color="black"           , muted_alpha=0.20, source=cds)
p2  = p.line("MONTH", "BRASIL" , line_alpha=0.25, width=2, color=Category20[14][ 0], muted_alpha=0.10, source=cds)
p3  = p.line("MONTH", "CANADA" , line_alpha=0.25, width=2, color=Category20[14][ 1], muted_alpha=0.10, source=cds)
p4  = p.line("MONTH", "CHILE"  , line_alpha=0.25, width=2, color=Category20[14][ 2], muted_alpha=0.10, source=cds)
p5  = p.line("MONTH", "EEUU"   , line_alpha=0.25, width=2, color=Category20[14][ 3], muted_alpha=0.10, source=cds)
p6  = p.line("MONTH", "MEXICO" , line_alpha=0.25, width=2, color=Category20[14][ 4], muted_alpha=0.10, source=cds)
p7  = p.line("MONTH", "URUGUAY", line_alpha=0.25, width=2, color=Category20[14][ 5], muted_alpha=0.10, source=cds)
p8  = p.line("MONTH", "CHINA"  , line_alpha=0.25, width=2, color=Category20[14][ 6], muted_alpha=0.10, source=cds)
p9  = p.line("MONTH", "INDIA"  , line_alpha=0.25, width=2, color=Category20[14][ 7], muted_alpha=0.10, source=cds)
p10 = p.line("MONTH", "JAPON"  , line_alpha=0.25, width=2, color=Category20[14][ 8], muted_alpha=0.10, source=cds)
p11 = p.line("MONTH", "UK"     , line_alpha=0.25, width=2, color=Category20[14][ 9], muted_alpha=0.10, source=cds)
p12 = p.line("MONTH", "SUIZA"  , line_alpha=0.25, width=2, color=Category20[14][10], muted_alpha=0.10, source=cds)
p13 = p.line("MONTH", "EURO"   , line_alpha=0.25, width=2, color=Category20[14][11], muted_alpha=0.10, source=cds)
p14 = p.line("MONTH", "VIETNAM", line_alpha=0.25, width=2, color=Category20[14][12], muted_alpha=0.10, source=cds)

p.add_tools(HoverTool(renderers=[p1], tooltips = [("ITCRM", "@ITCRM{0.0}")], mode="vline"))

#ARROWS
arrow_up   = Arrow(end=NormalHead(line_color='gray', line_width=1, line_alpha=0.50, size=15, fill_color='gray', fill_alpha=0.50),
                line_width = 10,
                line_color = 'gray',
                line_alpha = 0.50,
                x_start    = dt(2020, 1, 1).timestamp()*1000,
                x_end      = dt(2020, 1, 1).timestamp()*1000,
                y_start    = 310,
                y_end      = 380)

arrow_down = Arrow(end=NormalHead(line_color='gray', line_width=1, line_alpha=0.50, size=15, fill_color='gray', fill_alpha=0.50),
                line_width = 10,
                line_color = 'gray',
                line_alpha = 0.50,
                x_start    = dt(2020, 1, 1).timestamp()*1000,
                x_end      = dt(2020, 1, 1).timestamp()*1000,
                y_start    = 290,
                y_end      = 220)

depreciacion = Label(x=dt(2019, 8, 1).timestamp()*1000, y=330, text_font_size='10px', text_color='gray', text_align='right', text='DEPRECIACIÓN')
apreciacion  = Label(x=dt(2019, 8, 1).timestamp()*1000, y=255, text_font_size='10px', text_color='gray', text_align='right', text='APRECIACIÓN')

p.add_layout(arrow_up)
p.add_layout(arrow_down)
p.add_layout(depreciacion)
p.add_layout(apreciacion)

#LEGEND
legend = Legend(items=[
    ("ITCRM"         , [p1]),
    ("Brasil"        , [p2]),
    ("Canadá"        , [p3]),
    ("Chile"         , [p4]),
    ("Estados Unidos", [p5]),
    ("México"        , [p6]),
    ("Uruguay"       , [p7]),
    ("China"         , [p8]),
    ("India"         , [p9]),
    ("Japón"         , [p10]),
    ("Reino Unido"   , [p11]),
    ("Suiza"         , [p12]),
    ("Zona Euro"     , [p13]),
    ("Vietnam"       , [p14])], location="center", click_policy='mute')

p.add_layout(legend, 'right')

show(p)
```

```python
"|EXPORT .PNG FILE|"

export_png(p, filename="itcrm.png")
```

```python
"|# CREATE HTML FILE|"

output_file(filename="itcrm.html", title="ITCRM")
save(p)
```
