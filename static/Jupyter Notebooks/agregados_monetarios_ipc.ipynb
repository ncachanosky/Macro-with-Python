```python
"|IMPORT PACKAGES|"
import numpy              as np
import pandas             as pd
import datetime
from   bokeh.plotting     import show, figure, output_file, save
from   bokeh.io           import show, output_notebook, curdoc, export_png
from   bokeh.models       import ColumnDataSource,LinearAxis, Range1d, NumeralTickFormatter, LabelSet, Label, BoxAnnotation, DatetimeTickFormatter, Text, Span
from   bokeh.models.tools import HoverTool
from   bokeh.transform    import dodge
from   datetime           import datetime as dt
```

```python
"|IMPORT DATA|"
path = r'https://github.com/ncachanosky/research/blob/master/Economic%20Series/'
file = r'Resumen%20Estadistico%20-%20Argentina.xlsx?raw=true'
IO   = path + file

sheet = 'MONTHLY'

data = pd.read_excel(IO, sheet_name = sheet, usecols="B,CM:CO", skiprows=2, nrows=399, engine='openpyxl') # Be patient...

data = data.rename(columns={"M2.1":"M2"})
data = data.rename(columns={"IPC-C.1":"IPC"})
```

```python
"|CHECK DATA|"
data
```

```python
"|BUILD PLOT|"

cds = ColumnDataSource(data)

text1  = Label(x=25, y=325, x_units='screen', y_units='screen', text_font_size="12px", text_color="darkslategrey",
              text="VARIACIÓN ANAUL")

text2  = Label(x=25, y=310, x_units='screen', y_units='screen', text_font_size="12px", text_color="darkslategrey",
              text="Año          BM        M2        CPI")

text3  = Label(x=25, y=295, x_units='screen', y_units='screen', text_font_size="12px", text_color="darkslategrey",
              text="2007    24.0%   26.0%    25.7%")

text4  = Label(x=25, y=280, x_units='screen', y_units='screen', text_font_size="12px", text_color="darkslategrey",
              text="2008    10.2%   16.5%    23.0%")

text5  = Label(x=25, y=265, x_units='screen', y_units='screen', text_font_size="12px", text_color="darkslategrey",
              text="2009    11.9%   16.7%    14.8%")

text6  = Label(x=25, y=250, x_units='screen', y_units='screen', text_font_size="12px", text_color="darkslategrey",
              text="2010    31.0%   38.7%    25.3%")

text7  = Label(x=25, y=235, x_units='screen', y_units='screen', text_font_size="12px", text_color="darkslategrey",
              text="2011    39.0%   23.6%    22.7%")

text8  = Label(x=25, y=220, x_units='screen', y_units='screen', text_font_size="12px", text_color="darkslategrey",
              text="2012    37.9%   35.1%    25.2%")

text9  = Label(x=25, y=205, x_units='screen', y_units='screen', text_font_size="12px", text_color="darkslategrey",
              text="2013    22.7%   25.0%    27.7%")

text10 = Label(x=25, y=190, x_units='screen', y_units='screen', text_font_size="12px", text_color="darkslategrey",
              text="2014    22.6%   29.8%    38.4%")

text11 = Label(x=25, y=175, x_units='screen', y_units='screen', text_font_size="12px", text_color="darkslategrey",
              text="2015    34.9%   31.8%    27.9%")

text12 = Label(x=25, y=160, x_units='screen', y_units='screen', text_font_size="12px", text_color="darkslategrey",
              text="2016    31.7%   16.4%    40.7%")

text13 = Label(x=25, y=145, x_units='screen', y_units='screen', text_font_size="12px", text_color="darkslategrey",
              text="2017    21.8%   19.7%    24.7%")

text14 = Label(x=25, y=130, x_units='screen', y_units='screen', text_font_size="12px", text_color="darkslategrey",
              text="2018    40.7%   28.2%    47.6%")

text15 = Label(x=25, y=115, x_units='screen', y_units='screen', text_font_size="12px", text_color="darkslategrey",
              text="2019    34.5%   28.9%    54.0%")

text_anual1 = Label(x=300, y=325, x_units='screen', y_units='screen', text_font_size="12px", text_color="darkslategrey",
                    text="VARIACIÓN ANUALIZADA")

text_anual2 = Label(x=300, y=310, x_units='screen', y_units='screen', text_font_size="12px", text_color="darkslategrey",
                    text="BM  = 27.9%")

text_anual3 = Label(x=300, y=295, x_units='screen', y_units='screen', text_font_size="12px", text_color="darkslategrey",
                    text="M2  = 26.7%")

text_anual4 = Label(x=300, y=280, x_units='screen', y_units='screen', text_font_size="12px", text_color="darkslategrey",
                    text="IPC = 30.6%")

#BUILD FIGURE
p = figure(title = "EL HUB ECONÓMICO | EVOLUCIÓN BASE MONETARIA, M2, IPC (DIC-2016 = 100)",
           y_range      = (10, 8000),
           x_axis_label = "",
           y_axis_label = "",
           y_axis_type  = "log",
           plot_height  = 400,
           plot_width   = 700)

p.xaxis[0].formatter = DatetimeTickFormatter(months="%b %Y")
p.toolbar_location = "right"
p.toolbar.autohide = True

#AXIS 1 (LEFT)
#Inflación interanual
p.line("MONTH", "BM" , color="red"  , line_alpha=0.5, width=2, legend_label="BM" , source=cds, muted_alpha=0.20)
p.line("MONTH", "M2" , color="blue" , line_alpha=0.5, width=2, legend_label="M2" , source=cds, muted_alpha=0.20)
p.line("MONTH", "IPC", color="black", line_alpha=0.5, width=2, legend_label="IPC", source=cds, muted_alpha=0.20)


#ADD NOTES
p.add_layout(text1)
p.add_layout(text2)
p.add_layout(text3)
p.add_layout(text4)
p.add_layout(text5)
p.add_layout(text6)
p.add_layout(text7)
p.add_layout(text8)
p.add_layout(text9)
p.add_layout(text10)
p.add_layout(text11)
p.add_layout(text12)
p.add_layout(text13)
p.add_layout(text14)
p.add_layout(text15)

p.add_layout(text_anual1)
p.add_layout(text_anual2)
p.add_layout(text_anual3)
p.add_layout(text_anual4)

#LEGEND
p.xaxis[0].formatter  = DatetimeTickFormatter(months="%b %Y")
p.legend.location     = "bottom_right"
p.legend.orientation  = "horizontal"
p.legend.click_policy = "mute"
show(p)
```

```python
"|EXPORT .PNG FILE|"

export_png(p, filename="agregados_monetarios_ipc.png")
```
