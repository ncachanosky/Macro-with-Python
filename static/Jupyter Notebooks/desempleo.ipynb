```python
"|IMPORT PACKAGES|"
import numpy              as np
import pandas             as pd
import datetime
from   math               import pi
from   bokeh.plotting     import show, figure, output_file, save
from   bokeh.io           import show, output_notebook, curdoc, export_png
from   bokeh.models       import ColumnDataSource,LinearAxis, Range1d, NumeralTickFormatter, LabelSet, Label, BoxAnnotation, DatetimeTickFormatter, Text, Span
from   bokeh.models.tools import HoverTool
from   bokeh.models       import Arrow, NormalHead, OpenHead, VeeHead
from   bokeh.transform    import dodge
from   datetime           import datetime as dt
```

```python
"|IMPORT DATA|"
path = r'https://github.com/ncachanosky/research/blob/master/Economic%20Series/'
file = r'Resumen%20Estadistico%20-%20Argentina.xlsx?raw=true'
#file = r'Resumen%20Estadistico%20-%20Argentina.xlsx'
IO   = path + file

sheet = 'ANUAL'

data = pd.read_excel(IO, sheet_name = sheet, usecols="B,HT", skiprows=2, nrows=39, engine='openpyxl') # Be patient...

data['YEAR'] = data['YEAR'].dt.year
```

```python
"|CHECK DATA|"
data
```

```python
"|BUILD PLOT|"

cds = ColumnDataSource(data)

#Presidents' shades and names
Junta2      = BoxAnnotation(left=1976, right=1983, bottom=0, line_color="black", fill_color="green"  , fill_alpha=0.1)
Alfonsin    = BoxAnnotation(left=1983, right=1989, bottom=0, line_color="black", fill_color="red"    , fill_alpha=0.1)
Menem       = BoxAnnotation(left=1989, right=1999, bottom=0, line_color="black", fill_color="blue"   , fill_alpha=0.1)
DeLaRua     = BoxAnnotation(left=1999, right=2001, bottom=0, line_color="black", fill_color="red"    , fill_alpha=0.1)
Duhalde     = BoxAnnotation(left=2001, right=2003, bottom=0, line_color="black", fill_color="blue"   , fill_alpha=0.1)
NKirchner   = BoxAnnotation(left=2003, right=2007, bottom=0, line_color="black" ,fill_color="blue"   , fill_alpha=0.1)
CFK         = BoxAnnotation(left=2007, right=2015, bottom=0, line_color="black" ,fill_color="blue"   , fill_alpha=0.1)
Macri       = BoxAnnotation(left=2015, right=2019, bottom=0, line_color="black", fill_color="yellow" , fill_alpha=0.1)
Fernandez   = BoxAnnotation(left=2019, right=2021, bottom=0, line_color="black", fill_color="blue"   , fill_alpha=0.1)

Junta2_N    = Label(x=1982, y=1, text="Junta Militar", text_font_size="10pt", text_align="left", angle_units="deg", angle=90)
Alfonsin_N  = Label(x=1986, y=1, text="Alfonsín"     , text_font_size="10pt", text_align="center", text_baseline='bottom', angle_units="deg", angle= 0)
Menem_N     = Label(x=1994, y=1, text="Menem"        , text_font_size="10pt", text_align="center", text_baseline='bottom', angle_units="deg", angle= 0)
DeLaRua_N   = Label(x=2000, y=1, text="De la Rúa"    , text_font_size="10pt", text_align="left"  , text_baseline='middle', angle_units="deg", angle=90)
Duhalde_N   = Label(x=2002, y=1, text="Duhalde"      , text_font_size="10pt", text_align="left"  , text_baseline='middle', angle_units="deg", angle=90)
NKirchner_N = Label(x=2005, y=1, text="N. Kirchner"  , text_font_size="10pt", text_align="left"  , text_baseline='middle', angle_units="deg", angle=90)
CFK_N       = Label(x=2011, y=1, text="C. Kirchner"  , text_font_size="10pt", text_align="center", text_baseline='bottom', angle_units="deg", angle= 0)
Macri_N     = Label(x=2017, y=1, text="Macri"        , text_font_size="10pt", text_align="center", text_baseline='bottom', angle_units="deg", angle= 0)
Fernandez_N = Label(x=2020, y=1, text="Fernández"    , text_font_size="10pt", text_align="left"  , text_baseline='middle', angle_units="deg", angle=90)

#BUILD FIGURE
p = figure(title        = "EL HUB ECONÓMICO | TASA DE DESEMPLEO (PROMEDIO ANUAL)",
           y_range      = (0, 22),
           plot_height  = 400,
           plot_width   = 700)

p.toolbar_location = "right"
p.toolbar.autohide = True

p.add_tools(HoverTool(tooltips = [("Año","@YEAR"),("Desempleo", "@DESEMPLEO{0.0}")], mode='vline'))

#AXIS 1 (LEFT)
p.line('YEAR', 'DESEMPLEO', color='black', width=2, line_alpha=0.50, source=cds)
p.circle('YEAR', 'DESEMPLEO', color='black', alpha=0.50, size=7, source=cds)

#ADD PRSEIDENTS' SHADES
p.add_layout(Junta2)
p.add_layout(Alfonsin)
p.add_layout(Menem)
p.add_layout(DeLaRua)
p.add_layout(Duhalde)
p.add_layout(NKirchner)
p.add_layout(CFK)
p.add_layout(Macri)
p.add_layout(Fernandez)

#ADD PRSIDENTS' NAMES
p.add_layout(Junta2_N)
p.add_layout(Alfonsin_N)
p.add_layout(Menem_N)
p.add_layout(DeLaRua_N)
p.add_layout(Duhalde_N)
p.add_layout(NKirchner_N)
p.add_layout(CFK_N)
p.add_layout(Macri_N)
p.add_layout(Fernandez_N)

#LEGEND
p.xaxis.ticker = np.arange(1982,2022,2)
p.xaxis.major_label_orientation = pi/2
show(p)
```

```python
"|EXPORT .PNG FILE|"

export_png(p, filename="desempleo.png")
```

```python
"|CREATE HTML FILE|"

output_file(filename="desempleo.html", title="Tasa de desempleo")
save(p)
```
